# POST 練習

- 觀察 https://sea.cc.ntpu.edu.tw/pls/dev_stud/course_query_all.query_frame?flag=1 並撰寫爬蟲程式
- 抓取107-2法學院法學組課程
- 使用 requests + BeautifulSoup 實作
- 透過 pandas 輸出成 csv

```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://sea.cc.ntpu.edu.tw/pls/dev_stud/course_query_all.queryByAllConditions" 
```

```python
form_data ={
"qCollege":"法律學院".encode('big5'),  #用big5編碼後傳輸
"qdept":"LU31",
"qYear":107,
"qTerm":2,
"seq1":"A",
"seq2":"M"
}
```

```python
resp = requests.post(url, data=form_data)
soup = BeautifulSoup(resp.text, 'lxml')
```
